## Global Variable

- Global variables are declared outside of any function, usually at the top of the program or module.
- They can be accessed from any part of the program, including inside functions.

```python
global_var = 20  # global variable Accessable by any where

def my_function():
    print(global_var)

my_function()  # Output: 20
print(global_var)  # Output: 20
```

## Local Variable

- Local variables are declared inside a function or a specific code block.
- They are only accessible within the specific function where they are defined.

```python
local = 50

def my_function():
    global local  # if you want ot change the value of the global veriable from inside the function you use the {global veriable_name}
    local = 20
    local_var = 10  # local variable Accessable only inside the function

    print(local_var) 


my_function()  # Output: 10
print(local)
# print(local_var) # This would cause a NameError because we call it outside the function and the scope of this variable is only inside the function.
```
