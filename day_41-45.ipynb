### Short Hand If Else
- It is also a conditional statment but in a very simple and in very minimum form factor.
- They are use when we onlt test the code or for short form of conditional statment which have not that much or data.

```python
a = 1
b = 0
print("Max") if a>b else print("Min") 
```

```python
age = int(input("Enter your age : "))

print("Allowed") if age>18 else ("Not Allowed")
```

### Enumerate

- enumerate() function adds a counter to each item in a list or any other iterable, and returns a list of tuples containing the index position and the element for each element of the iterable.
- It turns the iterable into something we can loop through using indexes, where each item comes with its number (starting from 0 by default).

```python
marks = [15,23,24,12,21,25,28,21]

for i,mark in enumerate(marks): # here you see the i that work as a index that we want to calculate as far as the loop procced.
    print(mark)
    if i == 3:
        print("The lowest marks in the whole class.")
        
```
