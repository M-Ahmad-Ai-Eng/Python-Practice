## Exception Handling

Python Exception Handling allows a program to gracefully handle unexpected events (like invalid input or missing files) without crashing. Instead of terminating abruptly, Python lets you detect the problem, respond to it, and continue execution when possible.

```python
try:

        print("================================SELECT 2 NUMBERS TO ADD=============================")
        num_1 = int(input("Enter 1 number :"))
        num_2 = int(input("Enter 2 number :"))
        print( num_1 + num_2)
except ValueError:# this statement are execute when the mis match or un desirable value are feed into the
    print("Enter the Valid number")
# you also use multiple except block to handle different type of exception
except SyntaxError:
      print("Enter the correct value")#
```

```python
try:

        print("================================SELECT 2 NUMBERS TO Multiply=============================")
        num_1 = int(input("Enter 1 number :"))
        num_2 = int(input("Enter 2 number :"))
        print( num_1 * num_2)
except Exception as e:# this statement are execute when the mis match or un desirable value are feed into the
    print(e)# the e are used to print the actual error message and don't use the custome massage.And they are usefull for debugging purpose.to know the actual problem.
    print("Enter the Valid number")

print("================================PROGRAM ENDS HERE=============================")
```

### Finally keyword

- The finally keyword are used when we deal with the function rather then a normal program
- Because it is used inside the Function so that is why it return the value.
- Other wise we not use the traditional pyton print statment inside the function to verify is the function is working or raise any error.

## Without the Function

```python
try:
    list = [ 1 , 5 , 10 , 15 , 20 ]
    i = int(input("Enter the index of the number : "))
    print(list[i])
except:
    print("There is something Wrong here(Except)")
finally:
    print("The code is working all is good(Finally)")
```

## InSide the Function

```python
def func():
    try:
        list = [ 1 , 5 , 10 , 15 , 20 ]
        i = int(input("Enter the index of the number : "))
        print(list[i])

    except:
        return ("There is something Wrong here(Exept)")

    finally:
        return ("The code is working all is good(Finally)")

    print("Here is the print statment")# if you are intelligent to notice that the print stament are not execute single time inside the Function.
```

### Raising Custom Error
They help us to add some custom error to avoid the malfunctioning and some extra work in user side to take the user in a limited area not so.
They are help us when we want an restriction for something that user could do but we don't want.

```python
num = int(input("Enter any Even number : "))

if num % 2 != 0 :
    raise ValueError("Enter the Even Number")

print("You offically Enter the Even Number")
print(f"Which is : {num}")
```

```python
password = input("Enter your Password : ")
length = len(password)

if length <= 8 :
  raise ValueError("The password Character must be greater then 8.")
else:
  print(f"The password is correct : {password}")
```
