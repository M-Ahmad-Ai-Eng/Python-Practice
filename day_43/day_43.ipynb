### Virtual Environment
- The tool which use to Isolate the specific python environments in a single machine.
- Like they help to make the work on different projects with out any packages conflict and any problem.
- We use many tools or commands to make the Virtual Envirnment but the best one and recommened is the uv which is the project and package manager for python.
- You use any other tools but the uv is extremely fast and flexable and easy to setup.
- The venv is the build in module use to create a virtual env for your project.
- The uv the external one that you need to install by the **[powershell -ExecutionPolicy ByPass -c "irm https://astral.sh/uv/install.ps1 | iex"]** for windows.


```
uv init
```
- The commond help to initialize the folder to add the dependencies or packages to instal the packages like do the setup for you to add packages for your project.

```
uv add *package name*
```
- The commond help to add the pakages after **add** use the package name and thay will install it for you 
- If you not find the actual package name you simple go to <a href="https://pypi.org/">PyPi</a> and search for the package and they give the exact name of the package.

```
uv remove *package name*
```
- The commond help to remove the package from the virtual environment.

```
uv sync
```
- They install all the dependences of the file that comes with the project.
- If you add the **--upgrade** in the last they update all the packages or dependences on this project.
